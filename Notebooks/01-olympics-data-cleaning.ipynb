Supprimer les colones unitiles 

```python
medals = medals.drop(columns=["Unnamed: 0"])
results = results.drop(columns=["Unnamed: 0"])
```

normaliser les types 

athletes

```python
athletes["athlete_year_birth"] = athletes["athlete_year_birth"].astype("Int64")
```

host

```python
hosts["game_year"] = hosts["game_year"].astype(int)
```

extraire l’année partout

```python
medals["year"] = medals["slug_game"].str.extract(r"(\d{4})").astype(int)
results["year"] = results["slug_game"].str.extract(r"(\d{4})").astype(int)
```

créer la table propre pays → médailles

```python
medals["medal_id"] = (
    medals["slug_game"] + "_" +
    medals["discipline_title"] + "_" +
    medals["event_title"] + "_" +
    medals["medal_type"] + "_" +
    medals["country_name"]
)

medals_unique = medals.drop_duplicates(subset=["medal_id"])

# country_medals = (
#     medals_unique
#     .groupby(["year","country_name"])
#     .size()
#     .reset_index(name="medals")
# )
country_medals = (
    medals_unique
    .groupby(["year","country_name","medal_type"])
    .size()
    .reset_index(name="count")
)
```

```python
country_medals.head()
country_medals.tail()
country_medals.shape
```

Feature engineering

Trier par pays et année

```python
country_medals = country_medals.sort_values(["country_name","year"])
```

Créer historique cumulatif

```python
country_medals["past_medals"] = (
    country_medals
    .groupby("country_name")["count"]
    .cumsum()
    .shift(1)
    .fillna(0)
)
```

Créer moyenne mobile

```python
country_medals["avg_last_3"] = (
    country_medals
    .groupby("country_name")["count"]
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
country_medals["avg_last_3"] = country_medals["avg_last_3"].fillna(0)
```

Créer tendance

```python
country_medals["delta"] = (
    country_medals["count"] - 
    country_medals.groupby("country_name")["count"].shift(1)
).fillna(0)
```

```python
country_medals.to_csv("country_medals_ml.csv", index=False)
```
